## Working with Timestamp

```python
# creating a timestamp
timestamp = pd.Timestamp(2017, 1,1,12)
timestamp
```

```python
# creating a period
period = pd.Period('2017-01-01')
period
```

```python
# checking if given timestampt exist in the period
period.start_time < timestamp < period.end_time
```

```python
# converting timestamp to period
new_period = timestamp.to_period(freq='H')
new_period
```

```python
# converting period to timestamp
new_timestamp = period.to_timestamp(freq='H', how='start')
new_timestamp
```

## Finance and Statistics

```python
# percent change
stock['Change'] = stock.High.div(stock.High.shift())
stock['Change'].plot(figsize=(20,8))
```

```python
# stock returns
stock['Returns'] = stock['Change'].sub(1).mul(100)
stock['Returns'].plot(figsize=(20,8))
```

```python
# comparing two or more time series
ms_stock = pd.read_csv('./stock_data/MSFT_2006-01-01_to_2018-01-01.csv', index_col='Date', parse_dates=['Date'])
ms_stock.head()
```

```python
stock.High.plot()
ms_stock.High.plot()
plt.legend(['Google','Microsoft'])
plt.show()
```
