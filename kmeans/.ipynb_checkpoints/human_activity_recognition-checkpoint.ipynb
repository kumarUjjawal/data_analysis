# K-Means Clustering and PCA of Human Activity Recognition

```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import homogeneity_score, completeness_score, \
v_measure_score, adjusted_rand_score, adjusted_mutual_info_score, silhouette_score
%matplotlib inline
from IPython.display import display
np.random.seed(123)
```

```python
train = pd.read_csv('../data/activity_train.csv')
test = pd.read_csv('../data/activity_test.csv')
```

```python
# save labes as string
labels = train['activity']
train = train.drop(['rn','activity'], axis=1)
labels_keys = labels.unique().tolist()
labels = np.array(labels)
print('Label Keys: ',labels_keys)
```

```python
# check for missing values
temp = pd.DataFrame(train.isnull().sum())
temp.columns = ['Sum']
print('Amount of rows with missing values ' + str(len(temp.index[temp['Sum']>0])))
```

```python
# normalize the data
scalar = StandardScaler()
train = scalar.fit_transform(train)
```

```python
# finding optimal 'k' value
ks = range(1,10)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(train)
    inertias.append(model.inertia_)

plt.figure(figsize=(10,8))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
```

```python
def kmeans(n_clusters, data_frame, true_labels):
    kmeans = KMeans(n_clusters=n_clusters, random_state=123, n_init=30)
    kmeans.fit(data_frame)
    c_labels = kmeans.labels_
    df = pd.DataFrame({'clust_label': c_labels, 'orig_label': true_labels.tolist()})
    ct = pd.crosstab(df['clust_label'], df['orig_label'])
    y_clust = kmeans.predict(data_frame)
    display(ct)
    print('% 9s' % 'inertia  homo    compl   v-meas   ARI     AMI     silhouette')
    print('%i   %.3f   %.3f   %.3f   %.3f   %.3f    %.3f'
      %(kmeans.inertia_,
      homogeneity_score(true_labels, y_clust),
      completeness_score(true_labels, y_clust),
      v_measure_score(true_labels, y_clust),
      adjusted_rand_score(true_labels, y_clust),
      adjusted_mutual_info_score(true_labels, y_clust),
      silhouette_score(data_frame, y_clust, metric='euclidean')))
```

```python
kmeans(n_clusters=2, data_frame=train, true_labels=labels)
```

```python
# 6 clusters
kmeans(n_clusters=6, data_frame=train, true_labels=labels)
```
